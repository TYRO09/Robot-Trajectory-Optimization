#  Linear Joint-Space Trajectory

```python
import numpy as np
import matplotlib.pyplot as plt

# Total motion time
T = 5.0

# Time vector
t = np.linspace(0, T, 500)
```

```python
# Joint 1
q1_start = 0.3
q1_end = 1.0

# Joint 2
q2_start = -0.4
q2_end = 0.8
```

```python
q1_linear = q1_start + (q1_end - q1_start) * (t / T)
q2_linear = q2_start + (q2_end - q2_start) * (t / T)
```

```python
plt.figure(figsize=(8,4))

plt.plot(t, q1_linear, label="Joint 1")
plt.plot(t, q2_linear, label="Joint 2")

plt.xlabel("Time (s)")
plt.ylabel("Joint Angle (rad)")
plt.title("Linear Joint-Space Trajectory")
plt.legend()
plt.grid(True)

plt.show()
```

Both joint angles increase linearly with time from their start values to their end values. The plots are straight lines, indicating constant joint velocity throughout the motion. However, the velocity changes abruptly at the start and end of the trajectory, which can lead to jerky motion in real robots.

# Smooth Polynomial Trajectory

```python
T = 5.0
t = np.linspace(0, T, 500)

q1_start = 0.3
q1_end = 1.0

q2_start = -0.4
q2_end = 0.8
```

```python
s = t / T

q1_poly = q1_start + (q1_end - q1_start) * (3*s**2 - 2*s**3)
q2_poly = q2_start + (q2_end - q2_start) * (3*s**2 - 2*s**3)
```

```python
plt.figure(figsize=(8,4))

plt.plot(t, q1_poly, label="Joint 1")
plt.plot(t, q2_poly, label="Joint 2")

plt.xlabel("Time (s)")
plt.ylabel("Joint Angle (rad)")
plt.title("Smooth Polynomial Joint-Space Trajectory")
plt.legend()
plt.grid(True)

plt.show()
```

```python
dq1_poly = np.gradient(q1_poly, t)
dq2_poly = np.gradient(q2_poly, t)

plt.figure(figsize=(8,4))

plt.plot(t, dq1_poly, label="Joint 1 Velocity")
plt.plot(t, dq2_poly, label="Joint 2 Velocity")

plt.xlabel("Time (s)")
plt.ylabel("Joint Velocity (rad/s)")
plt.title("Joint Velocities for Polynomial Trajectory")
plt.legend()
plt.grid(True)

plt.show()
```

The polynomial trajectory starts and ends smoothly with zero velocity, unlike the linear trajectory. The joint angle curves are smooth near the boundaries, which reduces sudden changes in motion. This makes the polynomial trajectory more suitable for real robotic systems.

The linear joint-space trajectory changes joint angles at a constant rate, which results in abrupt changes in velocity at the start and end of motion. This behavior is visible in the joint angle plots as straight lines and implies infinite acceleration at the boundaries. In contrast, the polynomial trajectory starts and ends with zero velocity, producing smooth curves near the beginning and end of the motion. This smoothness reduces mechanical stress on actuators and minimizes vibrations. Linear trajectories are simple to compute but are not ideal for physical robots due to their discontinuous velocity profiles. Polynomial trajectories better respect real-world actuator limitations and are therefore more suitable for practical robotic applications.
